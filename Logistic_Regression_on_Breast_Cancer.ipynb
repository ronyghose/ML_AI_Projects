# **Logistic Regression on Breast Cancer**

**Problem Statement:** *Predict whether the cancer is benign or malignant.*


Dataset Link: "https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data?resource=download"

**In this project, we will see how to train a logistic regression model. This is intended as an introduction to logistic regression.**

We will be working with the Breast Cancer dataset, which contains some very detailed measurements of cells. Along with each observation of measurements, we have the diagnosis of the cell (malignant or not).

**Our goal is to train a model that will be able to predict whether or not a given cell is malignant given only its measurements.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import kagglehub

# 1. Download the dataset (already done in your code snippet)
path = kagglehub.dataset_download("uciml/breast-cancer-wisconsin-data")

print("Path to dataset directory:", path)

# 2. Determine the full path to the data file
# The downloaded content is usually placed directly in the directory 'path'.
# Check the filename inside the downloaded folder. For this dataset, it's often 'data.csv'.
file_name = 'data.csv'
full_file_path = os.path.join(path, file_name)

print("Full file path:", full_file_path)

# 3. Load the data into a pandas DataFrame
df = pd.read_csv(full_file_path)
```

# **Data Cleaning**



*   By using a heatmap, we can easily visualize the presence of NAs in the dataset and address them accordingly.
*   The 'ID' column, being irrelevant for analysis, will also be dropped.
*   Additionally, the target variable will be converted to numerical (1s and 0s) for model training.

Apart from these adjustments, the dataset appears clean and requires no further cleaning.





```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

We found no null values present in the data. So All set!

```python
df.drop(['Unnamed: 32', "id"], axis=1, inplace=True)
```

We dropped the irrelevant column "id' and "Unnamed: 32", since they dont have any role in our model setup.

```python
df.head()
```

So our columns is now dropped.

Here in thi data set "diagnosis" is target column, for which are building the model by the help of the rest of colmns as features.

So to train our model perfectly, we need to transform our target variable column "diagnosis" to numerical (1 or 0) column.



```python
df.diagnosis =[1 if value == "M" else 0 for value in df.diagnosis]
```

```python
df.head(5)
```

```python
df['diagnosis'] = df['diagnosis'].astype('category',copy=False)
plot = df['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0: Benign | 1: Malignant)")
fig = plot.get_figure()
```

# **Logistic Regression**

**Data Preprocessing for the Model**

The first thing to do is to do is to separate the target variable (here called "y") and the predictors (here called "X").

```python
y = df["diagnosis"] # our target variable
X = df.drop(["diagnosis"], axis=1) # our predictors
```

```python
X.head()
```

So we can notice that our each predictor column or features have values of different ranges and units. So make it even perfect for training we normalise it.

Also, there might be some outliers that could cause our model to perform badly.


Normalization could fix it.


**Normalization of the data**

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

# Fit the scaler to the data and transform the data
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

**Data Splitting: Training and Testing Sets**

Before training our model, we must split the full dataset into two parts: a Training Set (used to teach the model) and a Testing Set (used to evaluate the model's performance on unseen data). This prevents overfitting.

We have splitted our data set in 7:3 ratio for training and testing.


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# Creating logistic regression model
lr = LogisticRegression()

# Training the model on the training data
lr.fit(X_train, y_train)

# Predicting the target variable on the test data
y_pred = lr.predict(X_test)
```

```python
y_pred
```

# **Evaluate the Model**

So our model is trained and testing is also done.

Its time for the evaluation of the model i.e. how good our model is performing.

```python
from sklearn.metrics import accuracy_score

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

As you can see here, after training the model on the training set and evaluating its performance on the test set, we achieved a final accuracy of **0.98**.

This is a strong performance and indicates that the model is able to make accurate predictions on new, unseen data.

However, it's important to note that accuracy is just one measure of a model's performance, and it may not be the most appropriate measure for all problems. Depending on the problem and the specific requirements of the application, other metrics such as precision, recall, or F1 score may be more relevant.

So, we use the `classification_report` function from Scikit-learn to calculate those measures.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
